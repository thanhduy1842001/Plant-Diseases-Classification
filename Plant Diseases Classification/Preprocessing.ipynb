## Import library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import array_to_img
import cv2 as cv
```

```python
images_location = r"/content/drive/MyDrive/HW2---PaternRecognition/images/"
```

## Construct x_train

```python
x_train = []
width = 180
height = 180
dim = (width, height)

for i in range(0, 1821):
  img = cv.imread(f"{images_location}Train_{i}.jpg")
  img = cv.resize(img, dim)
  x_train.append(img)

x_train = np.array(x_train, dtype=np.float16) / 255.0
x_train
```

## Construct y_train 

```python
df = pd.read_csv(r"/content/drive/MyDrive/HW2---PaternRecognition/train.csv")
df.head()
```

```python
y_train = np.array([df['healthy'], df['multiple_diseases'], df['rust'], df['scab']])
y_train = np.argmax(y_train.T, axis=1)
```

## Save data

```python
import pickle
with open("/content/drive/MyDrive/HW2---PaternRecognition/train.pkl", "wb") as f:
    pickle.dump([x_train, y_train], f)
```
